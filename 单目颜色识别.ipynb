```python
import cv2
import numpy as np
```

```python
cap = cv2.VideoCapture(0)
ret = cap.set(3, 640)
ret = cap.set(4, 480)

def camera():
    while True:
        ret, frame = cap.read()
        # 转换成HSV空间
        hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
        # 设定绿色的阈值下限
        low_green = np.array([65, 43, 46])
        # 设定绿色的阈值上限
        up_green = np.array([80, 255, 255])
        # 设定蓝色的阈值下限
        low_blue = np.array([108, 160, 120])
        # 设定蓝色的阈值上限
        up_blue = np.array([113, 200, 170])
        # 设定红色的阈值下限
        low_red = np.array([2, 100, 100])
        # 设定红色的阈值上限
        up_red = np.array([180, 145, 255])
        
        # 做一个颜色层
        mask_green = cv2.inRange(hsv,low_green,up_green)
        mask_blue = cv2.inRange(hsv,low_blue,up_blue)
        mask_red = cv2.inRange(hsv, low_red, up_red)
        
        #确定小球的框框
        contourareas_green = cv2.findContours(mask_green.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)[-2]
        contourareas_blue = cv2.findContours(mask_blue.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)[-2]
        contourareas_red = cv2.findContours(mask_red.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)[-2]
        
        
        if len(contourareas_green)>0:
            #从最大的框框来确定小球的内切圆
            circle = max(contourareas_green,key=cv2.contourArea)
            #再从内切圆里面确定小球的圆心和半径
            ((x,y),z) = cv2.minEnclosingCircle(circle)
            x = int(x)
            y = int(y)
            z = int(z)
            print((x,y),'radius:%d' % z)
            if z >25:
                cv2.circle(frame,(x,y),z,(0,255,0),5)
        
        if len(contourareas_blue) >0:
            circle2 = max(contourareas_blue,key=cv2.contourArea)
            ((x2,y2),z2) = cv2.minEnclosingCircle(circle2)
            x2 = int(x2)
            y2 = int(y2)
            z2 = int(z2)
            print((x2,y2),'radius:%d' % z2)
            if z2 >25:
                cv2.circle(frame,(x2,y2),z2,(255,0,0),5)
                cv2.putText(frame, 'blue', ((x2+z2),(y2+z2)), cv2.FONT_HERSHEY_PLAIN, 2, (255,0,0), 2)
        
        if len(contourareas_red) >0:
            circle3 = max(contourareas_red,key=cv2.contourArea)
            ((x3,y3),z3) = cv2.minEnclosingCircle(circle3)
            x3 = int(x3)
            y3 = int(y3)
            z3 = int(z3)
            print((x3,y3),'radius:%d' % z3)
            if z3 >25:
                cv2.circle(frame,(x3,y3),z3,(0,255,0),5)
        cv2.imshow('camera',frame)
        key = cv2.waitKey(1) & 0xff
        if key == ord('q'):
            break
    cap.release()
    cv2.destroyAllWindows()
if __name__ == '__main__':
    camera()
```
